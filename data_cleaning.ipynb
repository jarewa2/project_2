```python
# Import libraries 
import pandas as pd
import numpy as np
from pathlib import Path
```

# Read all static csv files from Kaggle into dataframes

```python
# read csv files into dataframes
city_file = Path("weather_datasets/city_attributes.csv")
city_df = pd.read_csv(city_file)

hum_file = Path("weather_datasets/humidity.csv")
hum_df = pd.read_csv(hum_file)

pres_file = Path("weather_datasets/pressure.csv")
pres_df = pd.read_csv(pres_file)

temp_file = Path("weather_datasets/temperature.csv")
temp_df = pd.read_csv(temp_file)

wea_file = Path("weather_datasets/weather_description.csv")
wea_df = pd.read_csv(wea_file)

windir_file = Path("weather_datasets/wind_direction.csv")
windir_df = pd.read_csv(windir_file)

windsp_file = Path("weather_datasets/wind_speed.csv")
windsp_df = pd.read_csv(windsp_file)

city_df.head()
```

```python
# read all "Philadelphia" data into one dataframe             
philly_df = pd.DataFrame([hum_df.datetime, hum_df.Philadelphia,
                          pres_df.Philadelphia, temp_df.Philadelphia,
                          windir_df.Philadelphia, windsp_df.Philadelphia, wea_df.Philadelphia,]).transpose()

columns = ["datetime", "humidity", "pressure", "temperature", "wind-direction", "wind-speed", "weather-description"]
philly_df.columns = columns
philly_df.set_index(pd.to_datetime(philly_df['datetime'], infer_datetime_format=True), inplace=True)
philly_df.drop(columns=['datetime'], inplace=True)
philly_df.head(20)
```

```python
philly_df.dropna(inplace=True)
philly_df.isnull().sum()
```

```python
philly_df.to_csv("philly_weather.csv")
```
